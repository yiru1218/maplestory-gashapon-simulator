```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
import random
from flask import Flask
import random
```

```python
r = requests.get("https://tw-event.beanfun.com/MapleStory/eventad/EventAD.aspx?EventADID=8373") #將此頁面的HTML GET下來
# print(r.text) #印出HTML
# 將網頁資料以html.parser
soup = BeautifulSoup(r.text,"html.parser")
sel = soup.select("tbody tr td")

fashion_data = {}
fashion_data['道具名稱'] = []
fashion_data['機率'] = []

sel
```

```python
for idx, s in enumerate(sel):
    if(idx > 1 and idx % 2 == 0):
        fashion_data['道具名稱'].append(s.text.strip())
        #print(idx, s.text.strip()) # strip() 去除空格
    elif(idx > 1 and idx % 2 == 1):
        fashion_data['機率'].append(int(float(s.text.strip().replace('%', ''))*100))

df = pd.DataFrame(fashion_data)
df
# print(df.to_json(orient="table"))
pool = []
for item, num in zip(fashion_data['道具名稱'], df['機率']):
    pool += [item] * num
len(pool)
random.shuffle(pool)
pool
```

```python
def DrawTen(pool):
    res = []
    for i in range(10):
        res.append(random.choice(pool))
    return res
```

```python
@app.route("/fashion-data")
def getFashionBoxData():
    r = requests.get("https://tw-event.beanfun.com/MapleStory/eventad/EventAD.aspx?EventADID=8373") #將此頁面的HTML GET下來
    # print(r.text) #印出HTML
    # 將網頁資料以html.parser
    soup = BeautifulSoup(r.text,"html.parser")
    sel = soup.select("tbody tr td")

    fashion_data = {}
    fashion_data['道具名稱'] = []
    fashion_data['機率'] = []

    for idx, s in enumerate(sel):
        if(idx > 1 and idx % 2 == 0):
            fashion_data['道具名稱'].append(s.text.strip())
            #print(idx, s.text.strip()) # strip() 去除空格
        elif(idx > 1 and idx % 2 == 1):
            fashion_data['機率'].append(int(float(s.text.strip().replace('%', ''))*100))

    df = pd.DataFrame(fashion_data)
    fashion_data_json = df.to_json(orient="table")

    return fashion_data_json
```

```python
if __name__ == '__main__':
    app.run()
```
